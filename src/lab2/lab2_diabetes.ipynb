Imports

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

Dataset loading and exploration

```python
diabetes_dataset = pd.read_csv('../dataset/diabetes.csv')
diabetes_dataset.head()
```

```python
null_values = diabetes_dataset.isnull().sum()
print("Null values in each column:")
print(null_values)

zero_values = (diabetes_dataset == 0).sum()
print("Zero values in each column:")
print(zero_values)
```

Normalize & rescale dataset

```python
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(diabetes_dataset)
normalized_df = pd.DataFrame(normalized_data, columns=diabetes_dataset.columns)
print("Normalized & Scaled Dataset:")
print(normalized_df)
```

```python
# plot histograms of original and normalized data
import matplotlib.pyplot as plt
diabetes_dataset.hist(bins=15, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Original Data Histograms')
plt.show()
normalized_df.hist(bins=15, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Normalized Data Histograms')
plt.show()
```
