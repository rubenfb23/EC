Imports

```python
import pandas as pd
```

Dataset loading and exploration

```python
oil_spill_dataset = pd.read_csv('../dataset/oil_spill.csv')
unique_values = oil_spill_dataset.nunique()
print(unique_values)
```

Deleting colums that have only one unique value

```python
for column in oil_spill_dataset.columns:
    if oil_spill_dataset[column].nunique() <= 1:
        oil_spill_dataset = oil_spill_dataset.drop(column, axis=1)
        print(f"Column '{column}' has been deleted.")

print(oil_spill_dataset)
```

Detecting outliers and dealing with them

```python
dataset_outliers = oil_spill_dataset.copy()
Q1 = dataset_outliers.quantile(0.25)
Q3 = dataset_outliers.quantile(0.75)
IQR = Q3 - Q1
outliers = ((dataset_outliers < (Q1 - 1.5 * IQR)) | (dataset_outliers > (Q3 + 1.5 * IQR))).sum()
print("Outliers detected:")
print(outliers)
```

```python
#Deleting rows with outliers
dataset_no_outliers = dataset_outliers[~((dataset_outliers < (Q1 - 1.5 * IQR)) | (dataset_outliers > (Q3 + 1.5 * IQR))).any(axis=1)]
print("Dataset after removing outliers:")
print(dataset_no_outliers)
```
