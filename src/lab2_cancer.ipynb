```python
import pandas as pd
```

```python
cancer_dataset = pd.read_csv('../dataset/breast-cancer-Wisconsin.csv')
cancer_dataset.head()
```

```python
# clean and normalize data with scikit
from sklearn.preprocessing import MinMaxScaler
# check for null or zero values
null_values = cancer_dataset.isnull().sum()
print("Null values in each column:")
print(null_values)
zero_values = (cancer_dataset == 0).sum()
print("Zero values in each column:")
print(zero_values)
# drop rows with null values
cancer_dataset = cancer_dataset.dropna()
# replace zero values with mean of the column
for column in cancer_dataset.columns:
    if zero_values[column] > 0:
        mean_value = cancer_dataset[column].replace(0, pd.NA).mean()
        cancer_dataset[column] = cancer_dataset[column].replace(0, mean_value)
# normalize and scale data
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(cancer_dataset)
normalized_df = pd.DataFrame(normalized_data, columns=cancer_dataset.columns)
print("Normalized & Scaled Dataset:")
print(normalized_df)
```
