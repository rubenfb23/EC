```python
import pandas as pd
```

```python
cancer_dataset = pd.read_csv('../../dataset/breast-cancer-Wisconsin.csv')
cancer_dataset.info()
```

```python
# convert categorical data to numerical with scikit
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
cancer_dataset['diagnosis'] = label_encoder.fit_transform(cancer_dataset['diagnosis'])
cancer_dataset.info()
cancer_dataset.describe()
```

```python
# plot histograms of original and normalized data
import matplotlib.pyplot as plt
cancer_dataset.hist(bins=15, figsize=(20, 10), layout=(6, 6))
plt.suptitle('Original Data Histograms')
plt.show()
# Normalize & rescale dataset
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(cancer_dataset)
normalized_df = pd.DataFrame(normalized_data, columns=cancer_dataset.columns)
print("Normalized & Scaled Dataset:")
print(normalized_df)
normalized_df.hist(bins=15, figsize=(20, 10), layout=(6,
    6))
plt.suptitle('Normalized Data Histograms')
plt.show()
```
